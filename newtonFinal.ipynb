```python
import matplotlib.pyplot as plt
import numpy as np
import math
#Grace Horton
#Newton's Method

#f(x) = 1/1+e^x - 1/2
def f(x):
    return (1/(1+np.exp(x)))-(1/2)

#f'(x) = - e^x/(e^x+1)^2    
def fp(x):
    return -(np.exp(x)/((1+np.exp(x))**2)) 

#def fpp(x):
    #return -((np.exp(x)*(-np.exp(x)+1))/((np.exp(x)+1)**3))

#checking for convergence/divergence at root r = 0 
r = 0
#N(x) = x - f(x)/f'(x)
n = r-(f(r)/fp(r))
print('If Newtons Method converges, the answer should be zero. N(r)=',n)
print('===============================================================')

#boundaries for plotting [a,b]
#interval_left = -5.3
#interval_right = 5.3

interval_left = -2
interval_right = 2

#x values
x_val = np.linspace(interval_left, interval_right, num=10)
#y values 
y_val = f(x_val)

#list of starting points for looping
pointlist = x_val

plt.figure()

# plot of f(x)
plt.plot(x_val,y_val,'g-')

# plot of zero line for easy reference
plt.hlines(0, interval_left, interval_right)

#x and y axis labels
plt.xlabel("$x$", fontsize=16)
plt.ylabel("$f(x)$", fontsize=16)

#limit for x axis
plt.xlim((interval_left, interval_right))
#limit for y axis
plt.ylim((-2,9))

print('iter.:: xi ........:: |xi-xi-1|........:: |xi-xi-1|/|xi-1|...::')
print('===============================================================')
#pointlist loop
for x0 in pointlist:
    #g = gamma added as damping factor, for interval expansion
    #g = 0.1
    #precision and max iterations
    #tol = 1.e-4
    tol = 1.e-3
    i_max = 1000
    #initialized values for iteration loop
    reldiff = 1
    xi = x0
    itr = 0
    #iteration loop
    while reldiff > tol and itr < i_max:

        #print:: iteration, Distance from previous value, Relative distance btw approximations 
        print('%i, %15.12f, %15.12f, %15.12f' % (itr+1, x1, np.abs(f(xi)/fp(xi)), np.abs(f(xi)/fp(xi)/xi)))
        
        #relative difference
        reldiff = np.abs(f(xi)/fp(xi)/xi)
        #Newton's Iteration:: compute next xi
        x1 = xi - f(xi)/fp(xi)
        #x1 = xi - g*(f(xi)/fp(xi))
        #x1 becomes x0 and so on...
        xi = x1
        # increase counter
        itr += 1

    # plot for the number of necessary iterations at specific x0
    plt.plot(x0,itr,'r.', markersize=5)


plt.show()
#plt.close()
```
